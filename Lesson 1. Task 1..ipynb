Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
import json
from pprint import pprint
```

```python
main_link = 'https://api.github.com/users'
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
username = 'Varvara1991'
params = {'username': username}
```

```python
response = requests.get(f'{main_link}/{username}/repos')
```

```python
response.headers
```

```python
if response.ok:
    data = json.loads(response.text)
    pprint(data)
```

```python
if response.ok:
    data = json.loads(response.text)
result = []
for repository in data:
    result.append(repository.get('name'))
```

```python
print(f'Список репозиториев юзера {username}: \n {", ".join(result)}')
```

```python
with open('repo_list', 'w') as outfile:
    json.dump(result, outfile)
```
